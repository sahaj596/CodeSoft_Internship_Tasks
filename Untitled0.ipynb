```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Keep our Custom Sharpness Loss
def sharp_loss(y_true, y_pred):
    l1_loss = tf.reduce_mean(tf.abs(y_true - y_pred))
    ssim_loss = 1.0 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    return l1_loss + (0.5 * ssim_loss)

# 2. THE NEW U-NET ARCHITECTURE
def build_unet_autoencoder():
    inputs = layers.Input(shape=(128, 128, 4))

    # ENCODER (Compressing)
    c1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    # THE BOTTLENECK
    b = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p3)

    # DECODER WITH SKIP CONNECTIONS (Bridging c3, c2, c1 directly to the output)
    u1 = layers.UpSampling2D((2, 2))(b)
    concat1 = layers.Concatenate()([u1, c3]) # Skip Connection 1
    c4 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(concat1)

    u2 = layers.UpSampling2D((2, 2))(c4)
    concat2 = layers.Concatenate()([u2, c2]) # Skip Connection 2
    c5 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(concat2)

    u3 = layers.UpSampling2D((2, 2))(c5)
    concat3 = layers.Concatenate()([u3, c1]) # Skip Connection 3
    c6 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(concat3)

    outputs = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(c6)

    model = models.Model(inputs, outputs)
    model.compile(optimizer='adam', loss=sharp_loss)
    return model

print("U-Net Architecture Ready!")
```

```python
import glob
import cv2
import numpy as np
import random

# ==========================================
# 1. LOAD THE IMAGES
# ==========================================
image_paths = glob.glob('celeba/img_align_celeba/img_align_celeba/*.jpg')
subset_paths = image_paths[:5000]
clean_images = []

print("Loading and resizing images...")
for path in subset_paths:
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (128, 128))
    clean_images.append(img)
print(f"Successfully loaded {len(clean_images)} images.")

# ==========================================
# 2. DATA GENERATOR (Creates the masks)
# ==========================================
def get_masked_data(clean_images_array, mask_size=40):
    X_inputs, Y_targets = [], []
    for img in clean_images_array:
        h, w, _ = img.shape
        top, left = random.randint(0, h - mask_size), random.randint(0, w - mask_size)
        masked_img = img.copy()
        masked_img[top:top+mask_size, left:left+mask_size] = 0

        # 4th Attention Channel
        mask_channel = np.ones((h, w, 1), dtype=np.uint8) * 255
        mask_channel[top:top+mask_size, left:left+mask_size] = 0

        combined_input = np.concatenate([masked_img, mask_channel], axis=-1)
        X_inputs.append(combined_input)
        Y_targets.append(img)
    return np.array(X_inputs), np.array(Y_targets)

# ==========================================
# 3. TRAIN THE U-NET & SAVE IT
# ==========================================
print("Preparing mask-aware data pipeline...")
X_train, Y_train = get_masked_data(clean_images)
X_train = X_train.astype('float32') / 255.0
Y_train = Y_train.astype('float32') / 255.0

print("Building the U-Net Architecture...")
# This connects to the U-Net code we ran in the previous cell!
model = build_unet_autoencoder()

print("Training on Cloud GPU (15 Epochs)...")
history = model.fit(
    X_train, Y_train,
    epochs=15,
    batch_size=32,
    validation_split=0.1
)

# Overwrite the old brain file with the new U-Net brain
model.save('mask_aware_model.h5')
print("New U-Net Model saved successfully!")
```

```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
import numpy as np
import cv2
from PIL import Image

# Set up web page title and layout
st.set_page_config(page_title="AI Face Reconstruction", layout="centered")
st.title("🎭 AI Masked Face Reconstruction")
st.write("Upload a 4-channel image (RGB face with a black mask applied, plus an attention mask channel) to let the AI rebuild the missing pixels.")

# Load the trained model safely
@st.cache_resource
def load_ai_model():
    # We use compile=False because we only need the model for predictions, not training
    return tf.keras.models.load_model('mask_aware_model.h5', compile=False)

try:
    model = load_ai_model()
    st.success("🤖 AI Brain loaded successfully!")
except Exception as e:
    st.error(f"Could not load model file. Make sure 'mask_aware_model.h5' exists. Error: {e}")

# File Uploader on the website
uploaded_file = st.file_uploader("Choose a face image...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # Convert uploaded file to an OpenCV image
    file_bytes = np.asarray(bytearray(uploaded_file.read()), dtype=np.uint8)
    opencv_image = cv2.imdecode(file_bytes, 1)
    opencv_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2RGB)

    # Preprocess image to match what the model expects (128x128)
    original_resized = cv2.resize(opencv_image, (128, 128))

    # For demonstration on the website, we will automatically apply a 40x40 mask
    # so the user can see it work instantly on any normal photo they upload!
    h, w, _ = original_resized.shape
    masked_img = original_resized.copy()

    # Place a square mask right in the center for clean visualization
    top, left = (h // 2) - 20, (w // 2) - 20
    masked_img[top:top+40, left:left+40] = 0

    # Create the 4th channel (the binary attention mask)
    mask_channel = np.ones((h, w, 1), dtype=np.uint8) * 255
    mask_channel[top:top+40, left:left+40] = 0

    # Construct the 4-channel input tensor
    combined_input = np.concatenate([masked_img, mask_channel], axis=-1)
    input_tensor = np.expand_dims(combined_input, axis=0)
    input_tensor_normalized = input_tensor.astype('float32') / 255.0

    # Run prediction
    with st.spinner("AI is rebuilding the face details..."):
        predicted_output = model.predict(input_tensor_normalized)
        reconstructed_img = predicted_output[0]

    # Display the results side-by-side on the website
    col1, col2, col3 = st.columns(3)

    with col1:
        st.subheader("Your Upload")
        st.image(original_resized, use_container_width=True)

    with col2:
        st.subheader("Masked Input")
        st.image(masked_img, use_container_width=True)

    with col3:
        st.subheader("AI Result")
        st.image(reconstructed_img, use_container_width=True)
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
import cv2

st.set_page_config(page_title="True Face Unmasker", layout="centered")
st.title("🎭 Matrix-Powered Face Unmasker")
st.write("Using Linear Algebra to seamlessly remove masks without blurry squares.")

# File Uploader
uploaded_file = st.file_uploader("Upload your heavily masked image...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # 1. Load and resize the image
    file_bytes = np.asarray(bytearray(uploaded_file.read()), dtype=np.uint8)
    opencv_image = cv2.imdecode(file_bytes, 1)
    opencv_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2RGB)
    masked_img = cv2.resize(opencv_image, (128, 128))

    # 2. Detect the mask (Targeting the pure black pixels)
    gray = cv2.cvtColor(masked_img, cv2.COLOR_RGB2GRAY)

    # OpenCV's algorithm requires the missing area to be WHITE (255)
    inpaint_mask = np.zeros(gray.shape, dtype=np.uint8)
    inpaint_mask[gray <= 5] = 255

    # 3. THE MATRIX FIX (Telea Algorithm)
    # This solves the gradients from the edges inward, destroying the square artifact.
    with st.spinner("Solving matrix equations to seamlessly blend..."):
        reconstructed_img = cv2.inpaint(masked_img, inpaint_mask, 3, cv2.INPAINT_TELEA)

    # Display the results
    col1, col2 = st.columns(2)
    with col1:
        st.subheader("Your Uploaded Image")
        st.image(masked_img, use_container_width=True)
    with col2:
        st.subheader("Matrix Unmasked Result")
        st.image(reconstructed_img, use_container_width=True)
```

```python
%%writefile app.py
import streamlit as st
import torch
import numpy as np
import cv2
from PIL import Image
from diffusers import StableDiffusionInpaintPipeline
from streamlit_drawable_canvas import st_canvas

st.set_page_config(page_title="Simple AI Eraser", layout="wide")
st.title("🧹 Simple AI Eraser")

# 1. Load AI securely and quickly
@st.cache_resource
def load_ai_model():
    pipe = StableDiffusionInpaintPipeline.from_pretrained(
        "runwayml/stable-diffusion-inpainting",
        torch_dtype=torch.float16,
        safety_checker=None
    )
    pipe = pipe.to("cuda")
    return pipe

with st.spinner("Loading AI..."):
    pipe = load_ai_model()

if "final_image" not in st.session_state:
    st.session_state.final_image = None

# 2. Simple Interface
uploaded_file = st.file_uploader("Step 1: Upload Photo", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    raw_img = Image.open(uploaded_file).convert("RGB")
    img_resized = raw_img.resize((512, 512))

    col1, col2 = st.columns([1, 1])

    with col1:
        st.subheader("Step 2: Color over the object to delete it")
        canvas_result = st_canvas(
            fill_color="rgba(255, 0, 0, 0.5)",
            stroke_width=25,
            stroke_color="#FFFFFF",
            background_image=img_resized,
            update_streamlit=True,
            height=512,
            width=512,
            drawing_mode="freedraw",
            key="canvas",
        )

        st.subheader("Step 3: What should the AI draw in the empty space?")
        replacement_text = st.text_input("Example: 'man's mouth and chin' or 'seamless background'", value="man's mouth and chin, photorealistic")

    # 3. Process and Generate
    if canvas_result.image_data is not None:
        mask_np = cv2.cvtColor(canvas_result.image_data[:, :, 0:3], cv2.COLOR_RGB2GRAY)
        _, mask_np = cv2.threshold(mask_np, 1, 255, cv2.THRESH_BINARY)
        mask_img = Image.fromarray(mask_np).convert("RGB")

        with col2:
            if st.button("✨ Erase and Replace ✨"):
                if np.sum(mask_np == 255) < 100:
                    st.error("Please color over the mask first!")
                else:
                    with st.spinner("Working magic..."):
                        result = pipe(
                            prompt=replacement_text,
                            negative_prompt="ugly, blurry, artificial, mask, female, cartoon",
                            image=img_resized,
                            mask_image=mask_img,
                            num_inference_steps=25
                        ).images[0]
                        st.session_state.final_image = result

            if st.session_state.final_image is not None:
                st.image(st.session_state.final_image, caption="Final Result", use_column_width=True)
```
